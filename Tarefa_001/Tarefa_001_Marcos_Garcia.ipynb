RandomForestClassifier: É um dos modelos mais populares e robustos. Ele constrói várias árvores de decisão e combina suas previsões, o que o torna menos propenso a overfitting e geralmente mais preciso que uma única árvore ou a Regressão Logística.

GradientBoostingClassifier: Outro modelo de conjunto muito poderoso que constrói árvores de decisão sequencialmente, onde cada nova árvore tenta corrigir os erros das árvores anteriores. Ele tende a ter um desempenho excelente em muitos conjuntos de dados.

SVC (Support Vector Classifier): Máquinas de Vetores de Suporte podem ser muito eficazes, especialmente em dados com boa separabilidade. Com o uso de kernels (como o RBF), podem modelar relações não-lineares complexas, o que a Regressão Logística linear não consegue.

KNeighborsClassifier: Embora seja um modelo mais simples, baseado na distância entre os pontos de dados, pode ser uma melhoria significativa se os dados tiverem fronteiras de decisão não-lineares que a Regressão Logística não consegue capturar. É fácil de entender e implementar.

```python
# 4) Dividir em treino e teste (reprodutível e com estratificação do target)
RANDOM_STATE = 50
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.2,
    random_state=RANDOM_STATE,
    stratify=y
)

print("Treino:", X_train.shape, "Teste:", X_test.shape)
print("Proporção do target (treino):")
display(y_train.value_counts(normalize=True).rename("proporção"))
print("Proporção do target (teste):")
display(y_test.value_counts(normalize=True).rename("proporção"))
```

```python
def avaliar_classificacao(nome, y_true, y_pred):
    acc = accuracy_score(y_true, y_pred)
    cm = confusion_matrix(y_true, y_pred)

    print(f"\n=== {nome} ===")
    print(f"Accuracy: {acc:.4f}")
    print("\nClassification report:")
    print(classification_report(y_true, y_pred))
    print("Matriz de confusão:")
    print(cm)

    # Interpretação para classificação multiclasse
    print("\nInterpretação da Matriz de Confusão (Multiclasse):")
    print("- A diagonal principal mostra o número de predições corretas para cada classe.")
    print("- Os valores fora da diagonal principal mostram os erros do modelo, ou seja, onde o modelo previu uma classe diferente da verdadeira.")

    return {
        "acc": acc,
        "cm": cm
    }
```
